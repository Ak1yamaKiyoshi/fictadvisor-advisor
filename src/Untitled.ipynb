```python
import os
import openai
from dotenv import load_dotenv

load_dotenv()

api_key = os.getenv("OPENAI_API_KEY")
if api_key is None:
    raise ValueError("API key not found in environment variables")
    

openai.api_key = api_key

conversation_history = [
    {"role": "system", "content": "You are a frontend developer who refactores code"},
    {"role": "user", "content": f"{css_1_code}"},
    {"role": "assistant", "content": f"{sx_1_code}"},
    {"role": "user", "content": f"{css_2_code}"}
]

messages = [{"role": message["role"], "content": message["content"]} for message in conversation_history]


response = openai.ChatCompletion.create(
    model="gpt-4",
    messages=messages,
    max_tokens=4500  
)

assistant_reply = response['choices'][0]['message']['content']

print(assistant_reply)
```
